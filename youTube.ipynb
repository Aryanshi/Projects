```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
res = requests.get("https://www.youtube.com/")
soup = BeautifulSoup(res.text,"html.parser")
```

```python

records = []
like_list= []
dislikes_list= []
thumb_list= []
names_list= []
link_list = []
view_list = []

```

```python
for link in soup.find_all('a'):
    if str(link.get("href")).startswith('/watch?v='):
        link_list.append('https://www.youtube.com'+str(link.get("href")))
for views in soup.find_all('ul',attrs={'class':'yt-lockup-meta-info'}):
    view_list.append(views.text.split(' ')[0])
for name in soup.find_all('div',attrs= {'class':'yt-lockup-content'}):
    names_list.append(name.text.split('|')[0]) 
      
```

```python
for link in link_list[:10]:
    res1 = requests.get(link)
    soup1 = BeautifulSoup(res1.text,'html.parser')
    for like in soup1.find_all('button'):
        if str(like.get('aria-label')).startswith("like"):
            like_list.append(str(like.get('aria-label')))
            break 
    for dislike in soup1.find_all('button'):
        if str(dislike.get('aria-label')).startswith("dislike"):
            dislikes_list.append(str(dislike.get('aria-label')))
            break     
    for image in soup1.find_all('link',{'itemprop':'thumbnailUrl'}):
         thumb_list.append(str(image.get('href')))
```

```python
data_tup=list(zip(names_list,link_list[:10],thumb_list,view_list,like_list,dislikes_list))
df=pd.DataFrame(data_tup,columns="names links thumbnail views likes dislikes".split())
```
